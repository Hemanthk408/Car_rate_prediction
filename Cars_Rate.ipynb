```python

import pandas as pd


B=pd.read_json("/content/car_details_bengalur.json")
C=pd.read_json("/content/car_details_chennai.json")
N=pd.read_json("/content/car_details_New_Delhi.json")
J=pd.read_json("/content/car_details_jaipur.json")
P=pd.read_json("/content/car_details_pune.json")

pd.set_option("display.max_columns",None)

# Define the required columns
columns_to_keep = ['title', 'price', 'location', 'Registration Year', 'Insurance',
                   'Fuel Type','Kms Driven', 'RTO', 'Ownership']

# Select only required columns from each dataframe
df1 = B[columns_to_keep]
df2 = C[columns_to_keep]
df3 = N[columns_to_keep]
df4 = J[columns_to_keep]
df5=P[columns_to_keep]

# Merge all dataframes
merged_df = pd.concat([df1, df2, df3, df4,df5], ignore_index=True)

# Save the final dataframe
merged_df.to_csv('merged_data.csv', index=False)
```

```python
x=df.drop("price",axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=200,max_depth=52,random_state=0,min_samples_split=3).fit(x_train,y_train)
y_pred=model.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
model2=LinearRegression().fit(x_train,y_train)
y_pred=model2.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model3=ExtraTreesRegressor().fit(x_train,y_train)
y_pred=model3.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model6=ExtraTreesRegressor().fit(x_train,y_train)
y_pred=model6.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python

from sklearn.ensemble import RandomForestRegressor
model10 = RandomForestRegressor(n_estimators=100,criterion='squared_error',max_depth=2, random_state=0)
result=model10.fit(x_train, y_train)
y_pred = model10.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
#overfit to general -hypertune
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor(n_estimators=2,max_depth=12,random_state=22,min_samples_split=3).fit(x_train,y_train)
y_pred=model2.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
