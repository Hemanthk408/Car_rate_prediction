```python

import pandas as pd


B=pd.read_json("/content/car_details_bengalur.json")
C=pd.read_json("/content/car_details_chennai.json")
N=pd.read_json("/content/car_details_New_Delhi.json")
J=pd.read_json("/content/car_details_jaipur.json")


pd.set_option("display.max_columns",None)


# Load your tables (assuming they are CSVs)
B=pd.read_json("/content/car_details_bengalur.json")
C=pd.read_json("/content/car_details_chennai.json")
N=pd.read_json("/content/car_details_New_Delhi.json")
J=pd.read_json("/content/car_details_jaipur.json")

# Define the required columns
columns_to_keep = ['title', 'price', 'location', 'Registration Year', 'Insurance',
                   'Fuel Type','Kms Driven', 'RTO', 'Ownership']




# Select only required columns from each dataframe
df1 = B[columns_to_keep]
df2 = C[columns_to_keep]
df3 = N[columns_to_keep]
df4 = J[columns_to_keep]

# Merge all dataframes
merged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)

# Save the final dataframe
merged_df.to_csv('merged_data.csv', index=False)
```

```python
# Remove year from 'title'
df1['title'] = df1['title'].str.replace(r'\b\d{4}\b', '', regex=True).str.strip()

# Extract only the city name from 'location'
df1['location'] = df1['location'].str.split(',').str[1]

# Display the modified DataFrame
print(df1.head())
```
